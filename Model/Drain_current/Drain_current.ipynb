```python
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Drain Voltage with interval of 0.05V
Vd1 = []  
for  i in range(30):
    Vd1.append(i*0.05)
Vd = Vd1*6
#Source Volttage Vg = 0V
Vs = [-3]*30*6
Vg = []
# gate Voltage Vg = 0.2V to 1.2V with interval of 0.2V
for i in range(1,7):
    for j in range(30):
        Vg.append(i*0.2)
```

```python
inp1 = Vs +Vg + Vd
print((len(inp1)/3))
inp = np.transpose(np.array(inp1).reshape(3, int(len(inp1)/3)))
print(inp[3])
```

### 

```python
from keras.models import load_model
drain_current_p = load_model('drain_current_p.h5')
```

```python
Y_predict = drain_current_p.predict(inp)
Y_predict_1= [pow(10, element) for element in Y_predict]
```
