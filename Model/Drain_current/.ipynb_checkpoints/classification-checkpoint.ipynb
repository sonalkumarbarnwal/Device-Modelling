```python
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('output.csv')
#dataframe = pandas.read_csv("sonar.csv", header=None)
dataset = df.values
# split into input (X) and output (Y) variables
X = dataset[:,0:3].astype(float)
Y = dataset[:,11].reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1)
n_features = X_train.shape[1]
```

```python
model = Sequential()
model.add(Dense(5, input_dim=3, activation= 'relu'))
model.add(Dense(10, activation= 'relu'))
model.add(Dense(15, activation= 'relu'))
model.add(Dense(20, activation= 'relu'))
model.add(Dense(20, activation= 'tanh'))
model.add(Dense(20, activation= 'tanh'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=32, verbose=1)
```

```python
loss ,acurracy = model.evaluate(X_test, y_test, verbose=1)
print('MSE: %.3f' % (loss))
```

```python
y_predict = model.predict(X_train)
```

```python
y_predict1 = [1 if a_ > 0.5 else 0 for a_ in y_predict]
y_train1 = [1 if a_ > 0.5 else 0 for a_ in y_train]
```

```python
error = 0
for i in range(len(y_predict1)):
  if y_predict1[i] != y_train1[i]:
    print('false')
    print(i)
    error = error +1
print(error)
```
